```python
import pandas as pd 
import numpy as np
import math
import scipy

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import plotly.io as pio
pio.renderers.default = "plotly_mimetype+notebook"
```

```python
filepath = 'Data_Sets/owid-covid-data.csv'
df = pd.read_csv(filepath, delimiter=',')
df.set_index('location', inplace=True)
df.head()
Covid_data_raw_transport = df.loc[['China','United States','United Arab Emirates','Kenya','Russia','Netherlands']]
Covid_data_raw_passengers = df.loc[['United Kingdom','Spain','Germany','Italy','United States','Netherlands']]
Covid_data_raw_passengers
```

```python
fig = px.line(Covid_data_raw_passengers, x='date', y= 'new_cases_smoothed_per_million' ,color=Covid_data_raw_passengers.index,
                 range_x=['2020-02-01','2021-12-31'], range_y=[0,2500])

fig.show()
```

```python
fig = px.line(Covid_data_raw_transport, x='date', y= 'new_cases_smoothed_per_million' ,color=Covid_data_raw_transport.index,
                 range_x=['2020-02-01','2021-12-31'], range_y=[0,1500])

fig.show()
```

```python
fig = px.line(Covid_data_raw_passengers, x='date', y= 'people_fully_vaccinated_per_hundred' ,color=Covid_data_raw_passengers.index,
                 range_x=['2020-02-01','2022-07-31'], range_y=[0,100])

fig.show()
```
